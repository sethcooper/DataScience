### Linear models

```python
# diabetes dataset: 10 physiological variables, 442 patients, and disease progression 1 year later
diabetes = datasets.load_diabetes()
indices = np.random.permutation(len(diabetes.data))
diabetes_X_train = diabetes.data[indices[:-20]]
diabetes_X_test = diabetes.data[indices[-20:]]
diabetes_Y_train = diabetes.target[indices[:-20]]
diabetes_Y_test = diabetes.target[indices[-20:]]
```

```python
# Linear regression
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(diabetes_X_train,diabetes_Y_train)
print(regr.coef_)
# mean squared error
np.mean((regr.predict(diabetes_X_test)-diabetes_Y_test)**2)
```

```python
# variance score: 1 = 100% correlated
regr.score(diabetes_X_test,diabetes_Y_test)
```

#### Shrinkage

```python
X = np.c_[ .5, 1].T
Y = [.5, 1]
test = np.c_[ 0, 2].T
pyplot.figure()
# make some noise
for _ in range(6):
    this_X = 0.1*np.random.normal(size=(2,1))+X
    regr.fit(this_X,Y)
    pyplot.plot(test, regr.predict(test))
    pyplot.scatter(this_X,Y,s=3)
```

```python
# shrink the regression coefficients to zero; Ridge regression
regr = linear_model.Ridge(alpha=0.1)
pyplot.figure()
for _ in range(6):
    this_X = 0.1*np.random.normal(size=(2,1))+X
    regr.fit(this_X,Y)
    pyplot.plot(test, regr.predict(test))
    pyplot.scatter(this_X,Y,s=3)
```

```python
# we introduce a bias by the ridge regression which is a regularization
# can choose the alpha to minimize left out error on the diabetes dataset
alphas = np.logspace(-4, -1, 6)
from __future__ import print_function
print([regr.set_params(alpha=alpha
                      ).fit(diabetes_X_train, diabetes_Y_train,
                      ).score(diabetes_X_test, diabetes_Y_test) for alpha in alphas])
```

#### Sparsity

```python
# useful to select only informative features and drop others
# ridge regresison will penalize non-informative features but not set coefficients to zero
# Lasso (least absolute shrinkage and selection operator) can set some to zero: sparse method
regr = linear_model.Lasso()
scores = [regr.set_params(alpha=alpha
                         ).fit(diabetes_X_train,diabetes_Y_train
                         ).score(diabetes_X_test,diabetes_Y_test) for alpha in alphas]
best_alpha = alphas[scores.index(max(scores))]
regr.alpha = best_alpha
regr.fit(diabetes_X_train,diabetes_Y_train)
print(regr.coef_)
```

```python
# Exercise: Try classifying the digits dataset with nearest neighbors and a linear model.
#           Leave out the last 10% and use it to test prediction performance.


def GetAccuracy(real,predicted):
    numCorrect = 0
    for idx,realVal in enumerate(real):
        if predicted[idx]==realVal:
            #print('True')
            numCorrect+=1
    print('Accuracy=',numCorrect*1.0/len(real))

    
# nearest neighbors:
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
# split data randomly
digits_X = digits.data
digits_Y = digits.target
np.random.seed(0)
indices = np.random.permutation(len(digits_X))
splitPoint = int(0.9*len(indices))
digits_X_train = digits_X[indices[:splitPoint]]
digits_Y_train = digits_Y[indices[:splitPoint]]
digits_X_test = digits_X[indices[splitPoint:]]
digits_Y_test = digits_Y[indices[splitPoint:]]

knn.fit(digits_X_train,digits_Y_train)
prediction = knn.predict(digits_X_test)
print ('k-Nearest Neighbors')
GetAccuracy(digits_Y_test,prediction)
print ('Score:',knn.score(digits_X_test,digits_Y_test))

    
# linear model
# let's use sparsity and logistic regression
logistic = linear_model.LogisticRegression(C=1e2,penalty='l1') # let's lower the C to allow for more regularization
logistic.fit(digits_X_train,digits_Y_train)
prediction = logistic.predict(digits_X_test)
print('Logistic Regression')
GetAccuracy(digits_Y_test,prediction)
print ('Score:',logistic.score(digits_X_test,digits_Y_test))

# score is the same as my function above for accuracy
```

```python
# exercise
# Try classifying classes 1 and 2 from the iris dataset with SVMs, with the 2 first features.
# Leave out 10% of each class and test prediction performance on these observations.
# Warning: the classes are ordered, do not leave out the last 10%, you would be testing on only one class.
# Hint: You can use the decision_function method on a grid to get intuitions.



iris = datasets.load_iris()
X = iris.data
Y = iris.target

X = X[Y != 0, :2]
Y = Y[Y != 0]

from sklearn import svm
import matplotlib.pyplot as plt
svc = svm.SVC(kernel='linear')
np.random.seed(0)
indices = np.random.permutation(len(X))
splitPoint = int(0.9*len(indices))
X = X[indices]
Y = Y[indices]
trainX = X[:splitPoint]
trainY = Y[:splitPoint]
testX = X[splitPoint:]
testY = Y[splitPoint:]

svc.fit(trainX,trainY)
#print(svc.decision_function(trainX))
#print(svc.predict(trainX))

C = 1.0  # SVM regularization parameter
models = (svm.SVC(kernel='linear', C=C),
          #svm.LinearSVC(C=C),
          svm.SVC(kernel='rbf', gamma=0.7, C=C),
          svm.SVC(kernel='poly', degree=3, C=C))
models = (clf.fit(trainX, trainY) for clf in models)

# peek at solution
for num,model in enumerate(models):
    #model.predict(testX)
    print('Model:',model.kernel,'score:',model.score(testX,testY))
    plt.figure(num)
    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired,
                edgecolor='k', s=20)
    # test data has circles
    plt.scatter(testX[:, 0], testX[:, 1], s=80, facecolors='none',
                zorder=10, edgecolor='k')

    x_min = X[:, 0].min()
    x_max = X[:, 0].max()
    y_min = X[:, 1].min()
    y_max = X[:, 1].max()

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    #print('XX=',XX)
    #print('YY=',YY)
    #print('XX.ravel()=',XX.ravel())
    #print('np.c_[XX.ravel(), YY.ravel()]=',np.c_[XX.ravel(), YY.ravel()])
    Z = model.decision_function(np.c_[XX.ravel(), YY.ravel()])
    
    Z = Z.reshape(XX.shape)
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'],
                linestyles=['--', '-', '--'], levels=[-.5, 0, .5])

    plt.title(str(model.kernel))
plt.show()
```

```python
# exercise
# On the digits dataset, plot the cross-validation score of a SVC estimator with an linear kernel
#    as a function of parameter C (use a logarithmic grid of points, from 1 to 10).

import numpy as np
from sklearn.model_selection import cross_val_score, ShuffleSplit
from sklearn import datasets, svm
import matplotlib.pyplot as plt


digits = datasets.load_digits()
X = digits.data
y = digits.target

svc = svm.SVC(kernel='linear')
C_s = np.logspace(-10, 0, 10)

shuffleSplit = ShuffleSplit(3)

cvScores = list()
cvScoresStd = list()
for cVal in C_s:
    svc.C_ = cVal
    myScores = cross_val_score(svc, X_digits, y_digits, cv=shuffleSplit, n_jobs=-1)
    cvScores.append(np.mean(myScores))
    cvScoresStd.append(np.std(myScores))
    
plt.figure(1)
plt.clf()
plt.semilogx(C_s, cvScores)
plt.errorbar(C_s, cvScores, yerr=cvScoresStd)
#plt.semilogx(C_s, np.array(cvScores) + np.array(cvScoresStd), 'b--')
plt.ylabel('CV score')
plt.xlabel('Parameter C')
plt.show()
    
```

#### Grid search

```python
# can choose parameters automatically to maximize cross validation score
from sklearn.model_selection import GridSearchCV, cross_val_score

Cs = np.logspace(-6, -1, 10)
clf = GridSearchCV(estimator=svc, param_grid=dict(C=Cs),n_jobs=-1)
clf.fit(X_digits[:1000],y_digits[:1000])
print('best score:',clf.best_score_ )
print('best C value:',clf.best_estimator_.C)

print('score on test data:',clf.score(X_digits[1000:], y_digits[1000:]))
```
